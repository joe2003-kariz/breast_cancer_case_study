This notebook focuses on understanding the structure, quality, and relationships within the breast cancer dataset provided for the case study project. We will examine data types, missing values, feature distributions, correlations, and potential outliers to prepare for the modeling phase.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display and style
sns.set(style='whitegrid')
%matplotlib inline
```

## Load Dataset

```python
df = pd.read_csv('Data for Task 1.csv')
df.head()
```

## Basic Information About the Dataset

```python
df.info()
df.describe()
```

## Check for Missing Values

```python
df.isnull().sum()
```

## Drop Unnecessary Columns

```python
# Drop ID and empty column
if 'Unnamed: 32' in df.columns:
    df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
else:
    df.drop(columns=['id'], inplace=True)
```

## Encode Target Variable ('diagnosis')

```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

## Class Distribution

```python
sns.countplot(x='diagnosis', data=df)
plt.title('Distribution of Malignant and Benign Cases')
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.show()
```

## Feature Correlation Matrix

```python
plt.figure(figsize=(16, 12))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False, linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

## Pairplot of Selected Features

```python
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'diagnosis']
sns.pairplot(df[selected_features], hue='diagnosis', palette='husl')
plt.show()
```
